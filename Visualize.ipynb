```python
import matplotlib.pyplot as plt
import numpy as np

# Data
labels = ['Clinical Baseline\n(64 Channels)', 'Emotiv EPOC+\n(14 Channels)', 'Muse Headband\n(4 Channels)']
privacy_risk = [91.83, 95.17, 73.00]
utility_accuracy = [69.83, 66.83, 58.83]

x = np.arange(len(labels))
width = 0.35

fig, ax = plt.subplots(figsize=(10, 6))

# Create bars
rects1 = ax.bar(x - width/2, privacy_risk, width, label='Adversary ID Accuracy (Privacy Risk)', color='#d62728', edgecolor='black')
rects2 = ax.bar(x + width/2, utility_accuracy, width, label='Task Utility Accuracy', color='#1f77b4', edgecolor='black')

# Add horizontal baseline lines
ax.axhline(y=5.0, color='#d62728', linestyle='--', alpha=0.7, label='Random Chance (ID) - 5.0%')
ax.axhline(y=50.0, color='#1f77b4', linestyle='--', alpha=0.7, label='Random Chance (Task) - 50.0%')

# Add text, labels, and title
ax.set_ylabel('Accuracy (%)', fontsize=12)
ax.set_title('Privacy vs. Utility Tradeoff Across Spatial Resolutions', fontsize=14, pad=20)
ax.set_xticks(x)
ax.set_xticklabels(labels, fontsize=11)
ax.set_ylim(0, 110)

# Add exact percentage text on top of the bars
def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate(f'{height:.2f}%',
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom', fontsize=10)

autolabel(rects1)
autolabel(rects2)

ax.legend(loc='upper right', bbox_to_anchor=(1, 1.0), fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

plt.savefig('privacy_utility_tradeoff.png', dpi=300)
```
